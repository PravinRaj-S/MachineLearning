```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("priceList.xlsx")
df
```

```python
plt.scatter(df.area, df.price, color="red", marker='+')
plt.xlabel("Area (sqr ft)")
plt.ylabel("Price (US$)")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
plt.scatter(df.area, df.price, color="red", marker='+')
plt.xlabel("Area (sqr ft)")
plt.ylabel("Price (US$)")
plt.plot(df.area, reg.predict(df[["area"]]), color='b')
plt.show()
```

```python
reg.predict([[3300]])
```

```python
# Linear Equation
# Y = MX + B

# M => Coefficient
reg.coef_
```

```python
# X => Given area value
# B => Interception

reg.intercept_
```

```python
# Y = MX + B

reg.coef_ * 3300 + reg.intercept_
```

```python
df = pd.read_excel('predicPrice.xlsx')
df
```

```python
prices = reg.predict(df)
```

```python
df['price'] = prices
df['price'] = df['price'].astype(int)
df
```

```python
df.to_excel('predictedPriceList.xlsx', index=False)
```
